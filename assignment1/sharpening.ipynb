## 1. Filtering
### 1.2. Sharpening

```python
import cv2
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(20, 20))

# Original clean rice image (in Gray)
img = cv2.imread('./images/rice.png', cv2.IMREAD_GRAYSCALE)

# Sharpen rice image using Box Filter 5x5
img_box = cv2.boxFilter(img, -1, (5, 5))
img_box_sharp = cv2.add(img, cv2.subtract(img, img_box))

# Sharpen rice image using Gaussian Filter 5x5 with variance 2
img_gaussian = cv2.GaussianBlur(img, (5, 5), math.sqrt(2))
img_gaussian_sharp = cv2.add(img, cv2.subtract(img, img_gaussian))

plt.subplot(131), plt.imshow(img, cmap='gray')
plt.title('Original'), plt.axis('off')
plt.subplot(132), plt.imshow(img_box_sharp, cmap='gray')
plt.title('Sharpened using Box Filter 5x5 Blur'), plt.axis('off')
plt.subplot(133), plt.imshow(img_gaussian_sharp, cmap='gray')
plt.title('Sharpened using Gaussian Filter 5x5 Blur'), plt.axis('off')

# Display all images
plt.show()
```

```python
plt.figure(figsize=(20, 20))

# Original clean rice image (in Gray)
img = cv2.imread('./images/rice.png', cv2.IMREAD_GRAYSCALE)

# Sharpen rice image using Box Filter 5x5
img_box = cv2.boxFilter(img, -1, (5, 5))
img_box_details = cv2.subtract(img, img_box)

# Sharpen rice image using Gaussian Filter 5x5 with variance 2
img_gaussian = cv2.GaussianBlur(img, (5, 5), math.sqrt(2))
img_gaussian_details = cv2.subtract(img, img_gaussian)

plt.subplot(121), plt.imshow(img_box_details, cmap='gray')
plt.title('Details using Box Filter 5x5 Blur'), plt.axis('off')
plt.subplot(122), plt.imshow(img_gaussian_details, cmap='gray')
plt.title('Details using Gaussian Filter 5x5 Blur'), plt.axis('off')

# Display all images
plt.show()
```

We would expect the Box filter to give better results.

The Gaussian filter gives a higher weight to the pixels in the center allowing the resulting blurred image to preserve its sharper edges. As a result, when subtracting the blurred image from the original image to get the details image, we remove a lot of the sharpness around the edges of the rice. This leads the Gaussian filter into having a lesser sharpening effect on the original image.

On the other hand, the Box filter blurs the image evenly. As a result, when generating the details image we get a much blurrier image than that of the Gaussian filter, which makes it better at enhancing the sharpness of the original image. 

Also, as we can see from the two images above, the edges in the Gaussian filter details are much sharper than those in the Box filter details, which confirms the statements above.
