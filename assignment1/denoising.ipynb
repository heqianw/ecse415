```python
import cv2
import math
import numpy as np
import skimage.measure as skim
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(20, 20))

# Original clean peppers image
img = cv2.cvtColor(cv2.imread('./images/peppers.png'), cv2.COLOR_BGR2RGB)

# Peppers image with Gaussian noise
img_g_noise = cv2.cvtColor(cv2.imread('./images/gaussian_noise_peppers.png'), cv2.COLOR_BGR2RGB)
plt.subplot(131), plt.imshow(img_g_noise)
plt.title('Original with Gaussian Noise'), plt.axis('off')

# Filter noisy peppers image using Gaussian Filter 7x7 with 2 variance
img_gaussian = cv2.GaussianBlur(img_g_noise, (7, 7), math.sqrt(2)) 
plt.subplot(132), plt.imshow(img_gaussian)
plt.title('1. Gaussian Filter 7x7'), plt.axis('off')

# Filter noisy peppers image using Box Filter 7x7
img_box = cv2.boxFilter(img_g_noise, -1, (7, 7))
plt.subplot(133), plt.imshow(img_box)
plt.title('2. Box Filter 7x7'), plt.axis('off')

# Display all images
plt.show()

# Compare PSNRs of filtered images to that of the clean image
psnr_gaussian = skim.compare_psnr(img, img_gaussian)
psnr_box = skim.compare_psnr(img, img_box)
print ('Gaussian Filtered:', psnr_gaussian)
print ('Box Filtered:', psnr_box)
print ('Gaussian' if psnr_gaussian > psnr_box else 'Box', 'method provides better results since the PSNR of the filtered image is higher.')
```

```python
plt.figure(figsize=(20, 20))

# Peppers image with Salt and Pepper noise
img_sp_noise = cv2.cvtColor(cv2.imread('./images/saltandpepper_noise_peppers.png'), cv2.COLOR_BGR2RGB)
plt.subplot(131), plt.imshow(img_sp_noise)
plt.title('Original with Salt and Pepper Noise'), plt.axis('off')

# Filter noisy peppers image using Gaussian Filter 7x7 with 2 variance
img_gaussian = cv2.GaussianBlur(img_sp_noise, (7, 7), math.sqrt(2)) 
plt.subplot(132), plt.imshow(img_gaussian)
plt.title('4. Gaussian Filter 7x7'), plt.axis('off')

# Filter noisy peppers image using Median Filter 7x7
img_median = cv2.medianBlur(img_sp_noise, 7)
plt.subplot(133), plt.imshow(img_median)
plt.title('5. Median Filter 7x7'), plt.axis('off')

# Display all images
plt.show()

# Compare PSNRs of filtered images to that of the clean image
psnr_gaussian = skim.compare_psnr(img, img_gaussian)
psnr_median = skim.compare_psnr(img, img_median)
print ('Gaussian Filtered:', psnr_gaussian)
print ('Median Filtered:', psnr_median)
print ('Gaussian' if psnr_gaussian > psnr_median else 'Median', 'method provides better results since the PSNR of the filtered image is higher.')
```
