```python
import cv2
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(20, 20))

# Original amusent park image (in RGB) and clean Waldo image (in RGB)
img = cv2.cvtColor(cv2.imread('./images/where_is_waldo.jpg'), cv2.COLOR_BGR2RGB)
img_waldo = cv2.cvtColor(cv2.imread('./images/template1.png'), cv2.COLOR_BGR2RGB)
plt.subplot(321), plt.imshow(img)
plt.title('Original Amusement Park'), plt.axis('off')
plt.subplot(322), plt.imshow(img_waldo)
plt.title('Clean Waldo'), plt.axis('off')

# Scan amusement park image for clean Waldo template to create SSD map
ssd_map = cv2.matchTemplate(img, img_waldo, cv2.TM_SQDIFF)

# Find the point with the minimum SSD
_, _, min_loc, _ = cv2.minMaxLoc(ssd_map)

# Draw box around Waldo in original amusement park image and SSD map
h, w, _ = img_waldo.shape
bottom_right = (min_loc[0] + w, min_loc[1] + h)
color = (255, 0, 255)
thicc = 8
cv2.rectangle(img, min_loc, bottom_right, color, thicc)
cv2.rectangle(ssd_map, min_loc, bottom_right, color, thicc)

plt.subplot(323), plt.imshow(img)
plt.title('Amusement Park with Boxed Waldo'), plt.axis('off')
plt.subplot(324), plt.imshow(ssd_map, cmap='gray')
plt.title('SSD Map with Boxed Minimum SSD'), plt.axis('off')

# Display all images
plt.show()
```
